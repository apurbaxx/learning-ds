2. Mean Absolute Deviation

The mean absolute deviation of a dataset is the average distance between each point and the mean. It gives us an idea about the variability in a dataset

MAD = summation of (initial value of x - Mean of x)/N

Imp Note - Always chose data which is less spreaded

```python
sec_a = np.array([75,65,73,68,72,76])
sec_b = np.array([90,47,43,96,93,51])
no = np.array([1,2,3,4,5,6])
```

```python
mean = np.mean(sec_a)
```

```python
plt.figure(figsize=(10,3))
plt.scatter(sec_a,no,label="SEC A")
plt.scatter(sec_b,no, label="SEC B", color="red")
plt.plot([70,70,70,70,70,70], no, c="green", label="Mean")
plt.legend()
plt.show()
```

We will use Sec A data as they values provided are closer to mean value
We cant always afford to make this type of graph so we will use mean deviation method to decide which data to use

```python
mad_a = (np.sum(np.abs(sec_a-mean)))/len(sec_a)
mad_b = (np.sum(np.abs(sec_b-mean)))/len(sec_b)
```

```python
mad_a, mad_b
```

3. Standard Deviation

The standard deviation is a measure of the amount of variation or dispersion of a set of values. A low standard deviation indicates that the values tend to be close to the mean (also called expected value) of the set, while a high standard deviation indicates that the values are spread out over a wider range

```python
np.std(sec_a), np.std(sec_b)
```
