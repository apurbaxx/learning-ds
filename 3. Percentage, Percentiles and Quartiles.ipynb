Interquartile Range = 3rd Quartile - 1st Quartile

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("titanic.csv")
```

```python
dataset.head(3)
```

```python
dataset.isnull().sum()
```

If in case dataset contains missing values in age fill them with mean

```python
dataset["Age"].fillna(dataset["Age"].mean(), inplace=True)
```

```python
np.percentile(dataset["Age"], 25), np.percentile(dataset["Age"], 75), np.percentile(dataset["Age"], 100)
```

```python
dataset["Age"].min()
```

```python
dataset["Age"].max()
```

Median is basically 50 percentile

```python
np.percentile(dataset["Age"], 50)
```

```python
dataset["Age"].median()
```

```python
dataset.describe()
```

```python
sns.boxplot(x="Age", data=dataset)
plt.show()
```
