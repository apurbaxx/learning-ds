1. Skewnwess

- Skewness measures the asymetry of the distribution
- A skewness value of 0 indicates a perfectly symetrical distribution
- Positive Skewness indicates that the distribution is skewed to the right (i.e tail is longer on the right), while negetive skewness indicates a left skew (i.e the tail is longer on the left)

2. Frequency and Cumulative Distribution

- A frequency Distribution is a table that shows the number of occurrences of different values in a dataset
- A cumulative distribution shows the accumulation of frequencies upto a certain point. It is obtained by adding up the frequencies as you move through the values

For negatively skewed -> Mean < Median < Mode

For positively skewed -> Mode < Median <  Mean

for normal -> Everything in center of chart

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from seaborn import histplot
```

```python
dataset = pd.read_csv("titanic.csv")
```

```python
dataset.head(3)
```

```python
sns.histplot(x="Age", data=dataset)
plt.show()
```

```python
dataset["Age"].skew()
```

Since the skew is greater than 0 it is positively skewed

```python
data = np.random.normal(0,100,100)
```

```python
df = pd.DataFrame({"X": data})
```

```python
df["X"].skew()
```

Since the skew is less than 0 it is negatively skewed

```python
sns.histplot(x="X", data=df)
plt.show()
```
