Central Limit Theorem

The Central Limit Theorem states that when plotting a sample will be equal to the population mean and the sample distribution will approach normal distribution with variance equal to standarad error

There are few assumptions behind the CLT:
- The sample data must be sampled and selected randomly from the population
- There should not be any multicollinearity in the sampled data which is one sample should not influence other samples
- The sample size should not be more than 10 % of the population. Sample size greater than 30(n > 30) is not considered good.

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
pop_data = [np.random.randint(10,100) for i in range(10000)]
len(pop_data)
pop_table = pd.DataFrame({"Pop Data": pop_data})
```

```python
pop_table.head(3)
```

```python
plt.figure(figsize=(3,2))
sns.kdeplot(x="Pop Data", data = pop_table)
plt.show()
```

Kde Plot shows the probability density of our data. It basically tells which values occur more often

```python
sam_mean = []
for no_sample in range(50):
    sam_data = []
    for data in range(500):
        sam_data.append(np.random.choice(pop_data))
    sam_mean.append((np.mean(sam_data)))
    sam_data = []
```

```python
len(sam_mean)
```

```python
sample_m = pd.DataFrame({"Sample Mean": sam_mean})
```

```python
sample_m.head(3)
```

```python
plt.figure(figsize=(3,2))
sns.kdeplot(x="Sample Mean", data = sample_m)
plt.show()
```

```python
np.mean(pop_data)
```

```python
np.mean(sam_mean)
```

Hence the mean of sample mean and the population mean (pop data) is same
