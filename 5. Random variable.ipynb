1. Random variables

A random variable is a function that assigns a real number to each outcome in there sample space of a random experiment



- Discrete random variable

A random Variable that takes on a countable number of distinct values

- Continous Random Variable

A Random Variable that can take on any value within a given range or interval

2. Probability

Probability measures the likelihood of a particular outcome or the event occurring. It is typically expressed as a number between 0 and 1, where 0 indicates impossibility and 1 indicates certainty.
