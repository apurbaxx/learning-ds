```python
dataset = pd.read_csv("titanic.csv")
```

```python
dataset.head(3)
```

1. Using pd

```python
mn = dataset["Age"].mean()
```

2. Using np

```python
mn = np.mean(dataset["Age"])
```

```python
sns.histplot(x="Age", data=dataset, bins=[i for i in range(0, 81, 10)])
plt.plot([mn for i in range(0,300)],[i for i in range(0,300)], color= "red")
plt.show()
```

```python
dataset.isnull()
```

```python
md
```

```python
mnFare = dataset["Fare"].mean()
```

```python
mnFare
```

```python
sns.histplot(x="Fare", data=dataset, bins=[i for i in range(0, 81, 10)])
plt.plot([mn for i in range(0,300)], [i for i in range(0,300)], color="red")
plt.plot([md for i in range(0, 300)], [i for i in range(0, 300)], color="blue")
plt.show()
```

Mode

```python
 mo = dataset["Fare"].mode()[0]
```

```python
#Counts the number of time the value is repeated
dataset["Fare"].value_counts()
```

```python
mo
```

```python
sns.histplot(x="Fare", data=dataset, bins = [i for i in range(0,81,10)])
plt.plot([mn for i in range(0,300)], [i for i in range(0,300)], color="red", label="Mean")
plt.plot([md for i in range(0,300)], [i for i in range(0,300)], color="blue", label="Median")
plt.plot([mo for i in range(0,300)], [i for i in range(0,300)], color="green", label="Mode")
plt.legend()
plt.show()
```
